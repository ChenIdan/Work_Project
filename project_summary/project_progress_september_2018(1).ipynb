Another difference is the fact that the segal model is $\textbf{symmetric}$, and the kaplan model is not. 
The meaning is that if our DNA sequence that we want to predict the nucleusome occupancy on is
$$'CAC'$$
(let's say the nucleusome has 3 base pairs) than the kaplan model will calculate it like this:
$$p(C_{1})\cdot p(C_{1} | A_{2})\cdot p(A_{2} | C_{3}) $$
And the segal model will calculate it like this: 
$$p(A_{2} | C_{1}) \cdot p(A_{2})\cdot p(A_{2}|C_{3})$$
we tried to make our model symmetric by that that for every sequence given on the train for our program, to calculate the position weight matrix, we took also it's reverse as a sequence that were given by the train file.
I think that another way to solve this is like this:
first, when we get the fasta file, we generate a pwm matrix only from the lower half of the sequences given by the fasta file (each sequence is given also in reverse, so we cant say that we are ignoring from the upper halfs)
Now, lets say we generated our 74- rows lower half pwm. 
let's say that markov order is 1.
let's look on a uniform pwm matrix: 



```python
import numpy as np
import pandas as pd
from IPython.display import display

def to_pandas_dp(pwm):
    x_len = pwm.shape[1]
    y_len = pwm.shape[0]
    alphabet = ['A', 'C', 'G', 'T']
    cols = []
    for  i in range(x_len):
        tmp = ''.join(alphabet[i%4])
        
        tmp = alphabet[(i//4)%4] +alphabet[i%4] 
        cols.append( tmp)


    rows = range(0,y_len)

    for i in range(0,len(rows)):
        rows[i]=str(i)
    pwm_df = pd.DataFrame(data=pwm,
                  index=rows,
                  columns=cols)
    return pwm_df

    print display(pwm_df)
```

The problem with this method is that it ignores from the reverse reading order. If the segal model is calculating  the probability of sitting on the sequence 'CAC' as $$ p(C_{1}|A_{2}) \cdot p(A_{2})\cdot p(A_{2}|C_{1
3})$$ this matrix will bring to calculation 
like this: $$p(C_{1}|A_{2})\cdot p(A_{2})\cdot p(A_{2}|C_{3})$$
you can see it for yourself in this pwm matrix we generated. 

The soloution is to do 'fliping' on the upper half of the pwm matrix. In every row i , the value of
$p(A_{i-1} | C_{i})$ will be switched with the value representing $p(C_{i-1}| A_{i})$ and so on. 

The code will look like this: 

```python
def reverse_probs(up_pwm):
    mk_order = 1
    for row in up_pwm:
		for i in range(len(row)/2):
			# get current index on the current row to base 4 string representation with the right padding (this is the last parametr)
			num_str = np.base_repr(i,4, padding= mk_order+1 - len(str(np.base_repr(i,4,0))))
			#reverse the base 4 string representation
			num_str = num_str[::-1]

			if i<int(num_str,base=4): #this condition exists to prevent double swaping (AC to CA and CA to AC, for example)
				#swap sequence reading direction
				tmp_num = row[int(num_str, base=4)]
				row[int(num_str, base= 4)] = row[i]
				row[i] = tmp_num
```

Then we can do this while constructing our pwm: 

```python
low_pwm = np.load("low_pwm_example.npy")

#this is when we generate the pwm matrix:

up_pwm = np.flipud(low_pwm).copy() #flip the lower half of the pwm
reverse_probs(up_pwm)
pwm = np.vstack(( up_pwm,low_pwm[1:])) #stack the lower and the upper half verticly



to_pandas_dp(pwm)
```
